# Edge Filter

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
img=cv2.imread("E:/Dataset/Retinoblastoma.jpg",cv2.COLOR_BGR2GRAY)
#Apply filter for sharpening
edge_kernel=np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
edge_image=cv2.filter2D(src=img,ddepth=-1,kernel=edge_kernel)
plt.title("Convoluted image after applying edge filter")
plt.imshow(edge_image)
```

# Sharpen filter

```python
img=cv2.imread("E:/Dataset/Retinoblastoma.jpg",cv2.COLOR_BGR2GRAY)
#Apply filter for sharpening
sharpen_kernel=np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
sharpen_image=cv2.filter2D(src=img,ddepth=-1,kernel=sharpen_kernel)
plt.title("Convoluted image after applying edge filter")
plt.imshow(sharpen_image)
```
