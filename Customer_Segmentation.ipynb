```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("G:\\Dataset\\Mall_Customers.csv")
data.info()
```

#### The dataset does not contain null values

# Choose attributes for finding clusters

```python
X=data.iloc[:,[2,4]].values
```

# Find clusters

```python
from sklearn.cluster import KMeans
inertia=list()
for i in range(1,10):
    model=KMeans(n_clusters=i,init='k-means++',random_state=0)
    model.fit(X)
    inertia.append(model.inertia_)
    
```

```python
plt.plot(range(1,10),inertia)
plt.xlabel("Number of clusters")
plt.ylabel("Inertia Values")
plt.show()
```

#### From the above, it can be interpreted that the number of clusters are 4

# Model Training

```python
model=KMeans(n_clusters=4,init='k-means++',random_state=0)
y=model.fit_predict(X)#Compute cluster centers and predict cluster index
plt.scatter(X[y==0,0],X[y==0,1],c="red",label="Customer 1")
plt.scatter(X[y==1,0],X[y==1,1],c="blue",label="Customer 2")
plt.scatter(X[y==2,0],X[y==2,1],c="green",label="Customer 3")
plt.scatter(X[y==3,0],X[y==3,1],c="yellow",label="Customer 4")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```

#### From the above plot it can be interpreted that higher age group people have a lower spending score
